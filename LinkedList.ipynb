###**Single Linked List**


```
class Node:
  def __init__(self, data=None, next= None):
    self.data = data
    self.next = next
```

```
class LinkedList:
  #null linkedlist without any data
  def __init__(self):
    self.head = None
    self.size = 0

  def insert_first(self, data):
    self.size += 1
    node = Node(data, self.head)
    self.head = node
  
  def insert_last(self, data):
    if self.head is None:
      self.insert_first(data) # self.head = Node(data, None)
      return
    #if the next of pointer is not the end of list
    pointer = self.head
    while pointer.next: #next null mean last and the looping will stop at the last list
      pointer = pointer.next
    pointer.next = Node(data, None)
    self.size += 1

  def insert_after(self, index, data):
    if index < 0 or index >= self.size:
      raise Exception("Invalid Index")
    #insert after
    count, pointer = 0, self.head
    while pointer:
      if count == index:
        node = Node(data, pointer.next)
        pointer.next = node
        self.size +=1
        break
      pointer = pointer.next
      count += 1

  def insert_index(self, index, data):
    if index < 0 or index >= self.size:
      raise Exception("Invalid Index")
    #if the index is 0
    if index == 0:
      self.insert_first(data)
      return
    #insert on index location
    count, pointer = 0, self.head
    while pointer:
      if count == index-1:
        node = Node(data, pointer.next)
        pointer.next = node
        self.size +=1
        break
      pointer = pointer.next
      count += 1

  def insert_array_value(self, array_list):
    self.head = None #new list NULL
    for data in array_list:
      self.insert_last(data)
  
  def delete_first(self):
    if self.head is None:
      print("Empty List")
      return
    self.head = self.head.next
    self.size -= 1
  
  def delete_by_index(self, index):
    #check if data is null or index out of range (index start from 0)
    if index < 0 or  index >= self.size:
      raise Exception("Invalid Index")
    
    #if the first index
    if index == 0:
      self.delete_first() # self.head = self.head.next
      return
    
    #if the data between 2 data
    count,pointer = 0, self.head
    while pointer:
      if count == index-1:
        pointer.next = pointer.next.next  # A B C (delet B) so pointer A will point the C or A.next(B).next(C)
        self.size -= 1
        break
      pointer = pointer.next
      count += 1

  def delete_by_value(self, data):
    if self.head is None:
      print("Empty List")
      return

    if self.head.data == data:
      self.delete_first()
      return
    
    pointer = self.head
    while pointer.next:
      if pointer.next.data == data:
        pointer.next = pointer.next.next
        self.size -=1
        break
      pointer = pointer.next
    print("Value does not found")

  def print_list(self):
    #check is the linkedList in null
    if self.head is None:
      print("Empty List")
      return
    #if the list isn't empty
    pointer = self.head #pointer at the head/the first linkedlist
    ll_string =''
    while pointer:
      ll_string += str(pointer.data) + '-->'
      pointer = pointer.next
    print(ll_string)
```

###**Double Linked List**

```
class Node:
  def __init__(self, data=None, next=None, prev = None):
    self.data = data
    self.next = next
    self.prev = prev
```

```
class doubleLL:
  def __init__(self):
    self.head = None
    self.tail = None
    self.size = 0
  
  def insert_first(self, data):
    if self.head is None and self.tail is None:
      node = Node(data, None, None)
      self.head = node
      self.tail = node
    else:
      node = Node(data, self.head, None)
      self.head.prev = node
      self.head = node
    self.size += 1

  def insert_last(self, data):
    if self.head is None and self.tail is None:
      self.insert_first(data)
      return
    node = Node(data, None, self.tail)
    self.tail.next = node
    self.tail = node
    self.size += 1

  def insert_after(self, index, data):
    if index<0 or index>=self.size:
      raise Exception("Invalid Index")
    
    if index == self.size-1:
      self.insert_last(data)
      return
    
    pointer, counter = self.head, 0
    while pointer:
      if counter == index:
        node = Node(data, pointer.next, pointer)
        pointer.next.prev = node
        pointer.next = node
        break
      counter +=1
      pointer = pointer.next
    self.size +=1
  
  def delete_first(self):
    if self.size == 0:
      raise Exception("Empty Data")

    if self.head == self.tail:
      self.head = None
      self.tail = None
      self.size -= 1
      return
    
    self.head.next.prev = None
    self.head = self.head.next
    self.size -= 1
  
  def delete_last(self):
    if self.size == 0:
      raise Exception("Empty Data")

    if self.head == self.tail:
      self.delete_first()
      return
    
    self.tail.prev.next = None
    self.tail = self.tail.prev
    self.size -= 1

  def delete_after(self, index):
    if self.head is None and self.tail is None:
      print("Empty List")
      return

    if index<0 and index>=self.size:
      print("Index out ouf range")
      return

    if self.head == self.tail:
      print("List containts one element")
      return

    if index == self.size-1:
      print('This is the las list')
      return
    
    if index == self.size-2:
      self.delete_last()
      return

    pointer, counter = self.head, 0
    while pointer:  
      if index == counter:
        pointer.next.next.prev = pointer
        pointer.next = pointer.next.next
        self.size -= 1
        break
      counter += 1
      pointer = pointer.next

  def print_list(self):
    #if the list is empty
    if self.size == 0:
      print("Empty List")
      return

    pointer, ll_string = self.head, ''
    while pointer:
      ll_string += str(pointer.data) + '-->'
      pointer = pointer.next
    print(ll_string)
  
  def print_inverse(self):
    if self.size == 0:
      print("Empty List")
      return
    
    pointer, ll_string = self.tail, ''
    while pointer:
      ll_string += str(pointer.data) + '<--'
      pointer = pointer.prev
    print(ll_string)
```
